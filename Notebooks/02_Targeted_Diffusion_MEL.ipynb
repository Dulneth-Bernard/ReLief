```python
import os
import shutil

# 1. Define the Trap
base_dir = '/content/dataset/ISIC2018_Task3_Training_Input'
nested_trap = os.path.join(base_dir, 'ISIC2018_Task3_Training_Input')

# 2. Check and Fix
if os.path.exists(nested_trap):
    print(f"⚠️ Found the Double Folder Trap at: {nested_trap}")
    print("🚀 Moving images to the main folder... (This takes a few seconds)")

    # Move every file up one level
    files = os.listdir(nested_trap)
    for f in files:
        src = os.path.join(nested_trap, f)
        dst = os.path.join(base_dir, f)
        shutil.move(src, dst)

    # Delete the empty nested folder
    os.rmdir(nested_trap)
    print(f"✅ FIXED! Moved {len(files)} images. The nested folder is gone.")
else:
    print("✅ Folder structure looks correct (or nested folder already removed).")
```
